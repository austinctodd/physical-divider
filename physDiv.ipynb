# Physical Divider Simulation
This code simulates the detection of a physical divider in a road by cars that drive with various speeds and varying levels of error. The user should specify:
1. The road length
2. Physical divider locations
3. Number of cars (tuple of length N = number of different simulations)
4. GPS error
5. Accuracy of sensor error
